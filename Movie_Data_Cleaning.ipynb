```python
#Getting rid of rows with no budget information
df = df[df['budget']>0]
```

```python
len(df)
```

```python

#Calculate percentage return to account for wide variance over budget and gross. note that profit = gross-budget.
df['ROI_pct'] = 100*((df['gross']/df['budget'])-1)
```

```python
df.info()  #Memory usage shows up as just over half MB, which is small enough that we don't have to compress object types into category type
```

```python
#convert date stored as strings to ISO format
df['release_dt'] = pd.to_datetime(df['released'], infer_datetime_format=True)
```
